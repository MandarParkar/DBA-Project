```python
import pandas as pd
```

```python
df = pd.read_csv('./data/scores_output.csv')
df.info()
```

```python
df.head()
```

```python
tournament_df = df[['tournament', 'country']].drop_duplicates(ignore_index=True)
tournament_df.reset_index(inplace=True)
tournament_df.rename(columns={'index': 'tournament_pk'}, inplace=True)
tournament_df.info()
```

```python
tournament_df.head()
```

```python
teams_df = df[['home_team', 'tournament']].drop_duplicates(ignore_index=True)
teams_df.reset_index(inplace=True)
teams_df.rename(columns={'index': 'team_pk', 'home_team': 'team'}, inplace=True)
teams_df.info()
```

```python
teams_df.head()
```

```python
tournament_dict = tournament_df.set_index('tournament')['tournament_pk'].to_dict()
teams_df.replace({'tournament': tournament_dict}, inplace=True)
teams_df
```

```python
match_df = df[['home_team', 'away_team', 'tournament', 'year', 'month', 'day']].drop_duplicates(ignore_index=True)
match_df.reset_index(inplace=True)
match_df.rename(columns={'index': 'match_pk'}, inplace=True)
match_df.info()
```

```python
match_df.head()
```

```python
teams_dict = teams_df.set_index('team')['team_pk'].to_dict()
match_df.replace({'home_team': teams_dict, 'away_team': teams_dict, 'tournament': tournament_dict}, inplace=True)
match_df
```

```python
match_dict = match_df.set_index(['home_team', 'away_team', 'tournament', 'year', 'month', 'day'])['match_pk'].to_dict()
match_dict
```

```python
scores_df = df[['home_team', 'away_team', 'tournament', 'year', 'month', 'day', 'minute', 'home_score', 'away_score']]
scores_df.replace({'home_team': teams_dict, 'away_team': teams_dict, 'tournament': tournament_dict}, inplace=True)
scores_df['match_id'] = [match_dict[item] for item in list(zip(scores_df.home_team, scores_df.away_team, scores_df.tournament, scores_df.year, scores_df.month, scores_df.day))]
scores_df.drop(columns=['home_team', 'away_team', 'tournament', 'year', 'month', 'day'], inplace=True)
scores_df = scores_df[['match_id', 'minute', 'home_score', 'away_score']]
scores_df.reset_index(inplace=True)
scores_df
```
